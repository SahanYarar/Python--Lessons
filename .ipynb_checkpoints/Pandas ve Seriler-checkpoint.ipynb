```python
def calculateMean(df):
    totalSum = df.sum().sum()
    totalNum = df.size -df.isnull().sum().sum()
    return totalSum / totalNum
```

### Merge

```python
dataset1 = {
    "A" : ["A1","A2","A3"],
    "B" : ["B1","B2","B3"],
    "anahtar" : ["K1","K2","K3"]
}
dataset2 = {
    "X" : ["X1","X2","X3","X4"],
    "Y" : ["Y1","Y2","Y3","Y4"],
    "anahtar" : ["K1","K2","K5","K4"]
}
df1 = pd.DataFrame(dataset1,index = [1,2,3])
df2 = pd.DataFrame(dataset2,index = [1,2,3,4])
```

```python
df1
```

```python
df2
```

```python
#Merge ortak olan değerlere karşılık satırlar oluşturur
```

```python
#Join indexler üzerinden işlem yaparken merge column üzerinden işlemler gerçekleştiriyor
```

```python
pd.merge(df1,df2,on = "anahtar")
```

```python
import pandas as pd 
```

```python
youtube = pd.read_csv("Usvideos.csv")
```

```python
youtube
```

```python
list(youtube["likes"])
```

```python
def likesdislikes(likes,dislikes):
    likesList= []
    for key,value in list(likes.iteritems()):#İteritems fonksyonu Yanındaki indexle beraber alıyor
        likesList.append(value)
    dislikesList = []    
    for key,value in list(dislikes.iteritems()):
        dislikesList.append(value)  
    likedislike=list(zip(likesList,dislikesList))
    ratio = []
    for like,dislike in likedislike:
        if (like+dislike)==0:
            ratio.append(0)
        else:
            ratio.append(like / (like+dislike))
    return ratio       
            
        
        
youtube["likes_dislikes"]=likesdislikes(youtube["likes"],youtube["dislikes"])     

    
    
    

```

```python
youtube
```

```python
tot = youtube["likes"]+youtube["dislikes"]
ratio=youtube["likes"] / tot
youtube["ratio"]=ratio
youtube[["title","likes","dislikes","ratio"]].sort_values(by="ratio",ascending=False)
```
